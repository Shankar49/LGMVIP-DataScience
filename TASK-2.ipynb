### Image to Pencil Sketch with Python

```python
#importing required libraries
import cv2
import matplotlib.pyplot as plt
```

The following command can be used to read image using OpenCV.

```python
img=cv2.imread("dhoni.jpg")
```

It is not very convenient when new window opens up every time image has to be displayed and also the new window has to be deleted for further execution of code.
To avoid these hassles, we can use pyplot module of matplotlib library to display the image in the Jupyter notebook itself.

Convert BGR image to RGB

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.title("original image",size=18)
plt.axis(False)
plt.show()
```

**Step 1:** Convert to Grey Image
Using cvtColor function of OpenCV.

```python
grey_img=cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(grey_img)
plt.title("Grey image",size=18)
plt.axis(False)
plt.show()
```

**Step 2:** Invert Image
Inverting the image can be done by following methods

```python
invert_img = 255-grey_img
plt.imshow(invert_img)
plt.title("Inverted image",size=18)
plt.axis(False)
plt.show()
```

**Step 3:** Blur image
Apply Gaussian blur to the image.

```python
blur_img = cv2.GaussianBlur(invert_img, (21, 21), 0)
plt.imshow(blur_img)
plt.title("Blured image",size=18)
plt.axis(False)
plt.show()
```

**Step 4:** Invert Blurred Image

Repeat step 2

```python
invblur_img = 255 - blur_img
plt.imshow(invblur_img)
plt.title("inverted Blured image",size=18)
plt.axis(False)
plt.show()
```

**Step 5:** Sketch

* The sketch can be obtained by performing bit-wise division between the grayscale image and the inverted-blurred image.

```python
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
```

```python
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.title("Sketch image",size=18)
plt.axis(False)
plt.show()
```

**Original Image vs Sketch**

We can display the original image and sketch side by side for comparison.

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_img)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
