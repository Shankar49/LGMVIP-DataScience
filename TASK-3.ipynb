```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("terror_data.csv",encoding='latin1')
data.head()
```

```python
df = data.copy()
df.head()
```

```python
data_ext = data[['iyear','imonth','iday','country_txt','provstate','region_txt','city','latitude','longitude',"attacktype1_txt"
                 ,'nkill','nwound','target1','summary','gname','targtype1_txt','weaptype1_txt','motive']]
```

```python
data_ext.head()
```

```python
data_ext.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country'
                         ,'region_txt':'Region','provstate':'State','city':'City','longitude':'longitude','summary':'summary'
                         ,'attacktype1_txt':'Attacktype','targtype1_txt':'Target_type','weaptype1_txt':'Weapon','nkill':'kill'
                         ,'nwound':'Wound','target1':'Target'},inplace=True)
```

```python
data_ext.head()
```

```python
data_ext.shape
```

```python
data_ext.isnull().sum()
```

```python
data_ext['Wound'] = data_ext['Wound'].fillna(0)
data_ext['kill'] = data_ext['kill'].fillna(0)
```

```python
data_ext.info()
```

```python
data.describe()
```

```python
year = data_ext['Year'].unique()
years_count = data_ext['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year,y = years_count, palette = "coolwarm")
plt.xticks(rotation = 45)
plt.xlabel('Attacking Year',fontsize=10)
plt.ylabel('Number of Attacks per Year',fontsize=10)
plt.title('Attacks In Years',fontsize=20)
plt.show()
```

```python
pd.crosstab(data_ext.Year, data_ext.Region).plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Terrorist Activities By Region In Each Year',fontsize=20)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(data_ext['Country'].value_counts()[:10].index,data_ext['Country'].value_counts()[:10].values,palette='rocket')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation = 45)
plt.show()
```

```python
df = data_ext[['Year','kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
df.plot(kind='bar',alpha=0.7,ax=ax4,color="red")
plt.xticks(rotation = 50)
plt.title("People Died Due To Attack",fontsize=25)
plt.ylabel("Number of killed peope",fontsize=20)
plt.xlabel('Year',fontsize=20)

top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

```python
data_ext['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(10,5),color='lightcoral')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 10 Effected cities",fontsize=20)
plt.show()
```

```python
data_ext['Attacktype'].value_counts().plot(kind='bar',figsize=(10,10),color='lightblue')
plt.xticks(rotation = 25)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Name of attacktype",fontsize=15)
plt.show()
```

```python

data_ext[['Attacktype','kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(10,5),color=['firebrick'])
plt.xticks(rotation=50)
plt.title("Number of killed ",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

```python
data_ext[['Attacktype','Wound']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(10,5),color=['palegreen'])
plt.xticks(rotation=50)
plt.title("Number of people wounded ",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.countplot(data_ext["Target_type"],order=data_ext['Target_type'].value_counts().index,palette="icefire",edgecolor=sns.color_palette("Paired"));
plt.xticks(rotation=90)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.title("Attack per year",fontsize=20)
plt.show()
```

```python
data_ext['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='tomato',figsize=(20,10))
plt.title("Top 10 terrorist group attacks",fontsize=20)
plt.xticks(rotation=90)
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("Attack number",fontsize=15)
plt.show()
```

```python
data_ext[['Group','kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('kill',ascending=False).head(10).plot(kind='bar',color='crimson',figsize=(20,10))
plt.title("Top 10 terrorist group attack",fontsize=20)
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("No of killed people",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```
